```python
for column in df.columns:
    percentage = df[column].isnull().mean()
    print(column + ' ' + str(round(percentage * 100, 2)) + '%')
```

splitting the interest and site visited columns into multiple columns for better data reading

splitting interest column

```python
df['int_comma_count'] = df['Interests'].str.count(',')

max(df['int_comma_count'])

df['Interests'] = df['Interests'].str.strip("[]").str.replace("'","")
new_Columns = ['Interest1', 'Interest2','Interest3']
df[new_Columns] = df['Interests'].str.split(',', expand=True)
for col in new_Columns:
    if col in df:
        df[col] = df[col].fillna('')
        df[col] = df[col].str.strip()
```

Splitting the site visitted column

```python
df['sites_comma_count'] = df['Sites Visited'].str.count(',')

max(df['sites_comma_count'])

df['Sites Visited'] = df['Sites Visited'].str.strip("[]").str.replace("'","")
new_Columns = ['Sites1', 'Sites2','Sites3']
df[new_Columns] = df['Sites Visited'].str.split(',', expand=True)
for col in new_Columns:
    if col in df:
        df[col] = df[col].fillna('')
        df[col] = df[col].str.strip()
```

**OUTLIER** **DETECTION**

In the simplest way possible an outlier is basically the odd one out

and in this example we are looking for the outlire in the difference column

```python
columnName = 'Difference'

Q1 = df[columnName].quantile(0.25) # the value below 25% (our outlier)
Q3 = df[columnName].quantile(0.75) # the value above 75% (our outlier)
IQR = Q3 - Q1 #Calculating the Interquartile Range

outlier = df[((df[columnName] < (Q1 - 1.5 * IQR)) | (df[columnName] > (Q3 + 1.5 * IQR)))]
print(outlier)
```

 removing the rows from the df (Data Frame) that are also present in the outlier DataFrame

```python
df = df[~df.index.isin(outlier.index)]

df.head()
```
