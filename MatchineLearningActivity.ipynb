```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
G = nx.DiGraph()

# Add edges
G.add_edge('AB', 'Lib')
G.add_edge('Lib', 'ICT')
G.add_edge('ICT', 'EB')
G.add_edge('EB', 'Sci')
G.add_edge('Sci', 'C')
G.add_edge('C', 'A')
G.add_edge('A', 'B')
G.add_edge('C', 'Main')
G.add_edge('C', 'AB')
G.add_edge('Lib', 'SC')

# Draw with a cleaner layout
pos = nx.planar_layout(G)

plt.figure(figsize=(11, 6))
nx.draw(G, pos, with_labels=True, node_size=2400, arrows=True, arrowstyle='-|>', arrowsize=30, font_size=12, font_weight='bold')
plt.axis('off')
plt.show()
```

```python
# Define the number of vertices
num_vertices = 10

# Initialize the adjacency matrix with zeros
adj_matrix = np.zeros((num_vertices, num_vertices), dtype=int)

# Add edges to the graph
adj_matrix[0, 1] = 1
adj_matrix[6, 7] = 1
adj_matrix[1, 2] = 1
adj_matrix[2, 3] = 1
adj_matrix[3, 4] = 1
adj_matrix[4, 5] = 1
adj_matrix[5, 6] = 1
adj_matrix[5, 0] = 1
adj_matrix[5, 8] = 1
adj_matrix[1, 9] = 1

# Convert to pandas DataFrame for a clean table
nodes = list(G.nodes)
df = pd.DataFrame(adj_matrix, index=nodes, columns=nodes)
display(df)
```

 **CONCLUTION**

 By modeling the Smart Campus Shuttle System as a directed graph, we gain a clear view of how students and staff can navigate through key campus locations.

 The adjacency matrix gives a compact representation ideal for algorithmic analysis, while the adjacency list is memory-efficient and helpful for traversal operations like finding routes or detecting loops.

The directional nature of the graph reflects real-world traffic flow and time optimization. This model can support the future development of a machine learning system to predict shuttle wait times, suggest optimal paths, or manage real-time routing based on usage data.
